Considerando a base de dados experimento_test_ab.csv responda as
quest√µes abaixo:
1. Qual dos cen√°rios tem a maior taxa de convers√£o?
2. Calcule qual o tamanho da amostra necess√°ria para o desenvolvimento
de um teste A/B, seguindo os seguintes crit√©rios:
- a. O cen√°rio A, da base, como o inicial, que funciona hoje.
- b. Considere 95% de confian√ßa de que o efeito na convers√£o n√£o foi
aleat√≥rio com um n√≠vel de signific√¢ncia de 5% (alpha = 0,05).
- c. Tamb√©m considere 80% de certeza conseguir capturar o efeito
da nova abordagem.
- d. O aumento para 10% de convers√£o.


```python
# ==========================
# Quest√£o 1 - Qual dos cen√°rios tem a maior taxa de convers√£o?
# ==========================

print("\nüîπ Quest√£o 1: Qual dos cen√°rios tem a maior taxa de convers√£o?\n")

# Calcular taxa de convers√£o por vers√£o
conversion_rates = df.groupby('Vers√£o_P√°gina')['Convers√µes'].sum() / df.groupby('Vers√£o_P√°gina')['Visualiza√ß√µes'].sum()

print("Taxa de convers√£o por vers√£o:")
print(conversion_rates)

# Identificar cen√°rio com maior taxa
versao_maior = conversion_rates.idxmax()
print(f"\n‚úÖ A vers√£o com maior taxa de convers√£o √© a vers√£o: {versao_maior}")

# ==========================
# Quest√£o 2 - C√°lculo do tamanho de amostra para teste A/B
# ==========================

print("\nüîπ Quest√£o 2: C√°lculo do tamanho de amostra para teste A/B\n")

# ----- Letra a: Cen√°rio A como base atual -----
print("a) Usando a convers√£o da vers√£o A como base atual.")
p1 = conversion_rates['A']  # Convers√£o atual

# ----- Letra b: 95% de confian√ßa (alpha = 0.05) -----
print("b) Considerando 95% de confian√ßa (alpha = 0.05).")
alpha = 0.05

# ----- Letra c: 80% de poder estat√≠stico -----
print("c) Considerando 80% de poder estat√≠stico (power = 0.80).")
power = 0.80

# ----- Letra d: Aumento esperado para 10% de convers√£o -----
print("d) A meta de convers√£o √© 10% (p2 = 0.10).")
p2 = 0.10

# Tentar calcular o tamanho da amostra com os par√¢metros
try:
    effect_size = proportion_effectsize(p1, p2)
    analysis = NormalIndPower()
    sample_size = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha, ratio=1, alternative='larger')
    print(f"\nüìè Tamanho da amostra necess√°rio por grupo: {round(sample_size)} participantes")
except:
    print("\n‚ùå N√£o foi poss√≠vel calcular o tamanho da amostra.")
    print("A diferen√ßa entre p1 e p2 √© muito grande para o modelo estat√≠stico utilizado.")

# ==========================
# Extra - Simula√ß√£o com metas mais realistas
# ==========================
print("\nüîπ Simula√ß√£o extra: C√°lculo com metas mais pr√≥ximas da realidade\n")

possiveis_p2 = [0.001, 0.002, 0.005, 0.008, 0.01]  # Metas de convers√£o mais pr√≥ximas de p1

for p2_sim in possiveis_p2:
    try:
        effect_size = proportion_effectsize(p1, p2_sim)
        sample_size = analysis.solve_power(effect_size=effect_size, power=power, alpha=alpha, ratio=1, alternative='larger')
        print(f"‚û°Ô∏è  Para meta de {p2_sim:.2%} de convers√£o, amostra por grupo: {round(sample_size)}")
    except:
        print(f"‚ùå N√£o foi poss√≠vel calcular para p2 = {p2_sim:.2%}")
```

```python
# Quest√µes 3 a 10
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Montar o Google Drive (caso ainda n√£o tenha montado)
from google.colab import drive
drive.mount('/content/drive')

# Caminho do arquivo
file_path2 = '/content/drive/MyDrive/Bootcamp - WoMakersCode/pacientes.csv'
df = pd.read_csv(file_path2)

# QUEST√ÉO 3
print("\n--- Quest√£o 3 ---")
indices_q3 = [909, 751, 402, 400, 726, 39, 184, 269, 255, 769,
              209, 715, 677, 381, 793, 697, 89, 280, 232, 756, 358,
              36, 439, 768, 967, 699, 473, 222, 89, 639, 883, 558,
              757, 84, 907, 895, 217, 224, 311, 348, 146, 505, 273,
              957, 362]

amostra = df.loc[indices_q3, 'Idade']
media_amostra = amostra.mean()
print(f"M√©dia da amostra: {media_amostra:.2f}")

# Teste t unilateral (m√©dia > 50)
t_stat, p_val = stats.ttest_1samp(amostra, 50)
p_val_unilateral = p_val / 2 if t_stat > 0 else 1 - (p_val / 2)
print(f"t-stat√≠stica: {t_stat:.4f}, p-valor (unilateral): {p_val_unilateral:.4f}")
if p_val_unilateral < 0.05:
    print("Resultado: Rejeitamos H0. A m√©dia da amostra √© maior que 50.")
else:
    print("Resultado: N√£o rejeitamos H0. A m√©dia da amostra n√£o √© significativamente maior que 50.")

# QUEST√ÉO 4
print("\n--- Quest√£o 4 ---")
print("Resultado: Os grupos s√£o independentes, pois os indiv√≠duos com e sem condi√ß√µes adicionais s√£o distintos.")

# QUEST√ÉO 5
print("\n--- Quest√£o 5 ---")
indices_q5 = [690, 894, 67, 201, 364, 19, 60, 319, 588, 643, 855,
              623, 530, 174, 105, 693, 6, 462, 973, 607, 811, 346,
              354, 966, 943, 372]

subset = df.loc[indices_q5]
grupo1 = subset[subset['Estado_Saude'] == 1]['Pressao_Arterial']
grupo2 = subset[subset['Estado_Saude'] == 0]['Pressao_Arterial']

stat, p = stats.ttest_ind(grupo1, grupo2, equal_var=False)
print(f"Press√£o arterial m√©dia (condi√ß√µes): {grupo1.mean():.2f}")
print(f"Press√£o arterial m√©dia (sem condi√ß√µes): {grupo2.mean():.2f}")
print(f"t-stat√≠stica: {stat:.4f}, p-valor: {p:.4f}")
if p < 0.06:
    print("Resultado: Rejeitamos H0. H√° diferen√ßa significativa entre os grupos.")
else:
    print("Resultado: N√£o rejeitamos H0. Sem diferen√ßa significativa.")

# QUEST√ÉO 6
print("\n--- Quest√£o 6 ---")
anova = ols('Pressao_Arterial ~ C(Etnia)', data=df).fit()
anova_result = sm.stats.anova_lm(anova, typ=2)
print(anova_result)

if anova_result['PR(>F)'][0] < 0.05:
    print("Resultado: Rejeitamos H0. H√° diferen√ßa significativa entre os grupos √©tnicos.")
else:
    print("Resultado: N√£o rejeitamos H0. Sem diferen√ßa significativa.")

# QUEST√ÉO 7
print("\n--- Quest√£o 7 ---")
tabela_q7 = pd.crosstab(df['Genero'], df['Estado_Saude'])
chi2, p, dof, expected = stats.chi2_contingency(tabela_q7)
print(f"Qui-quadrado: {chi2:.4f}, p-valor: {p:.4f}")
if p < 0.05:
    print("Resultado: Rejeitamos H0. Existe rela√ß√£o entre g√™nero e condi√ß√µes de sa√∫de.")
else:
    print("Resultado: N√£o rejeitamos H0. Sem rela√ß√£o significativa.")

# QUEST√ÉO 8
print("\n--- Quest√£o 8 ---")
corr, p_valor = stats.pearsonr(df['Idade'], df['Pressao_Arterial'])
print(f"Correla√ß√£o de Pearson: {corr:.4f}, p-valor: {p_valor:.4f}")
if p_valor < 0.05:
    print("Resultado: Rejeitamos H0. Existe associa√ß√£o entre idade e press√£o arterial.")
else:
    print("Resultado: N√£o rejeitamos H0. Sem associa√ß√£o significativa.")

# QUEST√ÉO 9
print("\n--- Quest√£o 9 ---")
pa_cond = df[df['Estado_Saude'] == 1]['Pressao_Arterial']
media = pa_cond.mean()
std = pa_cond.std(ddof=1)
n = len(pa_cond)
conf_int = stats.t.interval(0.95, df=n-1, loc=media, scale=std/np.sqrt(n))
print(f"Intervalo de confian√ßa (95%) para press√£o arterial com condi√ß√µes de sa√∫de: {conf_int}")

# QUEST√ÉO 10
print("\n--- Quest√£o 10 ---")
amostra_pa = df['Pressao_Arterial'].sample(n=500, random_state=1) if len(df) > 500 else df['Pressao_Arterial']
stat, p = stats.shapiro(amostra_pa)
print(f"Shapiro-Wilk: estat√≠stica={stat:.4f}, p-valor={p:.4f}")
if p < 0.05:
    print("Resultado: Rejeitamos H0. A distribui√ß√£o n√£o √© normal.")
else:
    print("Resultado: N√£o rejeitamos H0. A distribui√ß√£o √© aproximadamente normal.")
```
