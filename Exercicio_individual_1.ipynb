```python
import pandas as pd

# Importando os dados
df = pd.read_csv('saude_do_sono_estilo_vida.csv')
df.head()
#print(df)
```

```python
# 1. Ao visualizar a base você percebeu que seria melhor alterar o nome de algumas colunas.
# a) Mude o ‘ID’ para ‘Identificador’
df = df.rename(columns={"ID": "Identificador"})

# b) Corrija o nome da coluna que indica a pressão sanguínea
df = df.rename(columns={"Pressão sanguíneaaaa": "Pressão sanguínea"})

# c) Mude a coluna ‘Ocupação’ para ‘Profissão’
df = df.rename(columns={"Ocupação": "Profissão"})

# d) A coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.
df = df.rename(columns={"Categoria BMI": "Categoria IMC"})

# Extra: A profissão "Vendedor(a) está representada por "Pessoa vendendora". Irei corrigir.
df["Profissão"] = df["Profissão"].replace("Pessoa Vendendora", "Vendedor(a)")

df.head()
```

```python
#2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mode]
horas_de_sono_por_profissao = df.groupby('Profissão')['Duração do sono'].agg(['mean', 'median', pd.Series.mode])
print(horas_de_sono_por_profissao)

#Em ordem decrescente:
horas_de_sono_por_profissao_ordem_decrescente = horas_de_sono_por_profissao.sort_values(by='mean', ascending=False)
print("\n", horas_de_sono_por_profissao_ordem_decrescente)

```

```python
# 3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?
engenheiros_software_obesos = df[(df["Profissão"] == "Eng. de Software") & (df["Categoria IMC"] == "Obesidade")]
total_engenheiros_software = df[df["Profissão"] == "Eng. de Software"]

porcentagem_engenheiros_software_obesos = (engenheiros_software_obesos.shape[0] / total_engenheiros_software.shape[0]) * 100
print("A porcentagem de obesos que atuam com engenharia de software é de", porcentagem_engenheiros_software_obesos, "%")
```

```python
# 4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)
profissoes = ["Advogado(a)", "Representante de Vendas"]
media_horas_sono = horas_de_sono_por_profissao[horas_de_sono_por_profissao.index.isin(profissoes)]['mean']

if(media_horas_sono['Advogado(a)'] < media_horas_sono['Representante de Vendas']):
    print("Advogar faz você dormir menos")
else:
    print("Ser representante de vendas faz você dormir menos")
```

```python
# 5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)
dados_filtrados = df[df["Profissão"].isin(["Enfermeiro(a)", "Médico(a)"])]
media_sono = dados_filtrados.groupby("Profissão")["Duração do sono"].mean()

if(media_sono["Enfermeiro(a)"] < media_sono["Médico(a)"]):
  print("Enfermeiro tem menos horas de sono")
else:
  print("Médico tem menos horas de sono")
```

```python
# 6. Faça um subconjunto com as colunas Identificador, Gênero, Idade, Pressão sanguínea e Frequência cardíaca.

df_subconjunto = df[["Identificador", "Gênero", "Idade", "Pressão sanguínea", "Frequência cardíaca"]]
print(df_subconjunto)
```

```python
# 7. Descubra qual a profissão menos frequente no conjunto. (Use value_counts)

profissoes_menos_frequentes = df["Profissão"].value_counts().tail(1)
print(profissoes_menos_frequentes)
```

```python
# 8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)
df[['Sistólica', 'Diastólica']] = df['Pressão sanguínea'].str.split('/', expand=True)

# Convertendo as colunas para inteiro
df['Sistólica'] = df['Sistólica'].astype(int)
df['Diastólica'] = df['Diastólica'].astype(int)

# Calculando a média dos dois valores para cada linha
df['Média'] = (df['Sistólica'] + df['Diastólica']) / 2

# Agrupando por Gênero e calculando a média da pressão
media_pressao_por_genero = df.groupby("Gênero")["Média"].mean()

if(media_pressao_por_genero["Homem"] > media_pressao_por_genero["Mulher"]):
  print("Homens tem maior pressão sanguínea média")
else:
  print("Mulheres tem maior pressão sanguínea média")
```

```python
# 9. É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?

moda_horas_sono = df["Duração do sono"].mode()[0]
if(moda_horas_sono == 8):
  print("É predominante entre os participantes dormir 8 horas por dia")
else:
  print("Não é predominante entre os participantes dormir 8 horas por dia")
```

```python
# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)
frequencia_cardiaca_acima_de_70 = df[df["Frequência cardíaca"] > 70]
media_passos_frequencia_cardiaca_acima_de_70 = frequencia_cardiaca_acima_de_70["Passos diários"].mean()
frequencia_cardiaca_menor_ou_igual_a_70 = df[df["Frequência cardíaca"] <= 70]
media_passos_frequencia_cardiaca_menor_ou_igual_a_70 = frequencia_cardiaca_menor_ou_igual_a_70["Passos diários"].mean()

if(media_passos_frequencia_cardiaca_acima_de_70 > media_passos_frequencia_cardiaca_menor_ou_igual_a_70):
  print("Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70")
else:
  print("Pessoas com frequências cardíacas acima de 70 NÃO dão mais passos que pessoas com frequência cardíaca menor ou igual a 70")

```
